## Scipy

```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
max_temp = [39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25]
```

```python
min_temp = [21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18]
```

```python
months = list(range(12))
```

```python
plt.scatter(months, min_temp, color='b', label='Minimun Temp')
plt.scatter(months, max_temp, color='r', label='Maximun Temp')
plt.legend()
plt.title('Temperatures in last year')
plt.xlabel('Month')
plt.ylabel('Temperature (\N{DEGREE SIGN}C)')
plt.tight_layout()
plt.show()
```

```python
from scipy import optimize
def yearly_temps(times, avg, ampl, time_offset):
    return (avg
            + ampl * np.cos((times + time_offset) * 2 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temps, months,
                                      max_temp)
res_min, cov_min = optimize.curve_fit(yearly_temps, months,
                                      min_temp)
```

```python
days = np.linspace(0, 12, num=365)

plt.figure()
plt.scatter(months, max_temp, color='r')
plt.plot(days, yearly_temps(days, *res_max), color='r', label='Max Temp')
plt.scatter(months, min_temp, color='b')
plt.plot(days, yearly_temps(days, *res_min), color='b', label='Min Temp')
plt.title('Temperatures in last year')
plt.xlabel('Month')
plt.ylabel('Temperature (\N{DEGREE SIGN}C)')
plt.legend()
plt.show()
```

## Matplotlib

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
```

```python
titanic = pd.read_csv(url)
```

```python
titanic.columns
```

```python
male = len(titanic[titanic['sex'] == 'male']['sex'])
female = len(titanic[titanic['sex'] == 'female']['sex'])
```

```python
sex = [male, female]
```

```python
sex
```

```python
plt.figure(figsize=(6,6))
plt.pie(sex, labels=['male', 'female'], autopct='%1.1f', wedgeprops = {'edgecolor': 'black'})
plt.title('Male/Female proportion chart')
plt.tight_layout()
plt.show()
```

```python
titanic.columns
```

```python
males = titanic[titanic['sex'] == 'male'][['age', 'fare']]
females = titanic[titanic['sex'] == 'female'][['age', 'fare']]
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(10,10))
plt.scatter(females['age'], females['fare'], color='g', label='Females')
plt.scatter(males['age'], males['fare'], color='r', label='Males')
plt.yticks(np.arange(0,550,50))
plt.legend()
plt.tight_layout()
plt.show()
```
